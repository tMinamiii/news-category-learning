```python
%matplotlib inline
import vectorize.learning_data as ld
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
news_vecs = ld.load('ldata/2017-10-14.svdtd') 
tuid = ld.load('ldata/2017-10-14.tuid')
cat_list = list(tuid.categories)
label = news_vecs[:,0]
data = news_vecs[:,1]
CATEGORIES = {'IT総合':'IT', '映画':'Cinema', '経済総合':'Economics', '野球':'Baseball',
              '社会':'Social', 'ライフ総合':'Life', 'エンタメ総合':'Entertaiment', 'サッカー':'Football', 'スポーツ総合':'Sports'}
```

```python
mat_dic = {}
for l, d in zip(label,data):
    ind = l.index(1)
    total = None 
    if ind in mat_dic:
        total = mat_dic[ind]
    
    if total is None:
        total = d
    else:
        total += d
    mat_dic.update({ind:total})
```

```python
plt.rcParams['figure.figsize'] = (7.0,5.0)
d_len = len(data[0])
for k,v in mat_dic.items():
    ave = v/d_len
    x = [i for i in range(0,d_len)]
    plt.plot(x, ave[0:], label=CATEGORIES[cat_list[k]])
plt.legend()
```

```python
d_len = len(data[0])
for k,v in mat_dic.items():
    ave = v/d_len
    x = [i for i in range(0,50)]
    plt.plot(x, ave[0:50], label=CATEGORIES[cat_list[k]])
plt.legend()
```

```python
for k,v in mat_dic.items():
    ave = v/d_len
    x = [i for i in range(449,d_len-2)]
    plt.plot(x, ave[449:d_len-2], label=CATEGORIES[cat_list[k]])
plt.legend()
```
